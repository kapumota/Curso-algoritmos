## 1.1 Variables

Antes de llegar a la definición de variables, relacionémoslas con antiguas ecuaciones matemáticas. Todos hemos resuelto muchas ecuaciones matemáticas desde la infancia. Como ejemplo, considera la siguiente ecuación:

$$
x^2 + 2y - 2 = 1
$$

No tenemos que preocuparnos por el uso de esta ecuación. Lo importante aquí es notar que la ecuación tiene **nombres** ($x$ e $y$) que **contienen valores** (datos). Esos nombres (x e y) son marcadores de posición para representar datos. De manera similar, en la programación  necesitamos algo que contenga datos, y las **variables** son la manera de hacerlo.

---

## 1.2 Tipos de datos

En la ecuación mencionada anteriormente, las variables $x$ e $y$ pueden tomar distintos valores, por ejemplo, números enteros (10, 20), números reales (0.23, 5.5), o simplemente 0 y 1. Para resolver la ecuación, necesitamos relacionarlos con el tipo de valores que pueden tomar, y **tipo de dato** es el nombre que se usa en programación para este propósito.  

Un tipo de dato en un lenguaje de programación es un conjunto de datos con valores predefinidos. Ejemplos de tipos de datos son: entero (integer), punto flotante (floating point), número sin signo (unsigned number), carácter (character), cadena (string), etc.

La memoria de la computadora está llena de ceros y unos. Si tenemos un problema y queremos codificarlo, sería muy difícil dar la solución en términos de ceros y unos. Para ayudar a los usuarios, los lenguajes de programación y los compiladores nos proporcionan **tipos de datos**. Por ejemplo, un entero ocupa 2 bytes (el valor exacto depende del compilador), un float ocupa 4 bytes, etc. Esto indica que en memoria combinamos 2 bytes (16 bits) y lo llamamos un entero; combinamos 4 bytes (32 bits) y lo llamamos un float. Un tipo de dato reduce el esfuerzo de codificación.  
A alto nivel, existen dos tipos de datos:

1. **Tipos de datos definidos por el sistema** (también llamados **tipos de datos primitivos**).  
2. **Tipos de datos definidos por el usuario**.

### Tipos de datos definidos por el sistema (primitivos)

Los tipos de datos que están definidos por el sistema se llaman **tipos primitivos**. Los tipos de datos primitivos más comunes en muchos lenguajes de programación son: `int`, `float`, `char`, `double`, `bool`, etc. El número de bits asignado a cada tipo de dato primitivo depende del lenguaje de programación, del compilador y del sistema operativo. Para el **mismo** tipo de dato primitivo, distintos lenguajes pueden utilizar tamaños diferentes. Dependiendo del tamaño de los tipos de datos, el dominio de valores disponibles también variará.  
Por ejemplo, "int" puede ocupar 2 bytes o 4 bytes. Si ocupa 2 bytes (16 bits), entonces el rango de valores es de -32,768 a +32,767 $(-2^{15}$ a $2^{15}-1)$. Si ocupa 4 bytes (32 bits), el rango va de -2,147,483,648 a +2,147,483,647 $(-2^{31}$ a $2^{31}-1)$. Lo mismo ocurre con otros tipos de datos.

### Tipos de datos definidos por el usuario

Si los tipos de datos definidos por el sistema no son suficientes, la mayoría de los lenguajes de programación permiten a los usuarios **definir sus propios tipos de datos**, llamados **tipos de datos definidos por el usuario**. Ejemplos comunes son las **estructuras** en C/C++ y las **clases** en Java.  

Por ejemplo, en el siguiente fragmento de código, combinamos varios tipos de datos del sistema y llamamos al nuevo tipo de dato definido por el usuario **"newType"**. Esto da más flexibilidad y comodidad para manejar la memoria de la computadora:

```c
struct newType {
    int data1;
    float data2;
    char data;
};
```

## 1.3 Estructuras de datos

Una vez que tenemos datos en variables, necesitamos algún mecanismo para manipular esos datos y **resolver problemas**. Una **estructura de datos** es una forma particular de almacenar y organizar datos en una computadora de modo que puedan usarse de manera eficiente. En términos generales, una estructura de datos es una **fórmula especial** para organizar y almacenar datos. Los tipos de estructura de datos más comunes incluyen: arreglos, archivos, listas enlazadas, pilas, colas, árboles, grafos, etc.

Dependiendo de cómo se organicen los elementos, las estructuras de datos se clasifican en dos tipos:

1. **Estructuras de datos lineales**: se accede a los elementos en orden secuencial, pero no es obligatorio que todos los elementos se almacenen de manera contigua. Ejemplos: listas enlazadas, pilas y colas.  
2. **Estructuras de datos no lineales**: los elementos se almacenan/acceden en un orden no lineal. Ejemplos: árboles y grafos.

---

## 1.4 Tipos de datos abstractos (ADTs)

Antes de definir los tipos de datos abstractos, consideremos otra visión de los tipos de datos definidos por el sistema.  
Sabemos que, por defecto, todos los **tipos de datos primitivos** (int, float, etc.) soportan operaciones básicas como la suma y la resta, y el sistema proporciona las implementaciones para ellos. Para los **tipos de datos definidos por el usuario**, también necesitamos definir operaciones. La implementación de estas operaciones puede hacerse cuando realmente deseemos usarlas. En general, los tipos de datos definidos por el usuario se **definen junto con sus operaciones**.

Para simplificar el proceso de **resolver problemas**, combinamos las estructuras de datos con sus operaciones y los llamamos **Tipos de Datos Abstractos (ADTs)**. Un ADT consta de dos partes:

1. Declaración de datos.  
2. Declaración de operaciones.

Los ADTs comúnmente utilizados incluyen: listas enlazadas, pilas, colas, colas de prioridad, árboles binarios, diccionarios, conjuntos disjuntos (Unión y Búsqueda), tablas hash, grafos y muchos otros.  
Por ejemplo, la **pila** (stack) usa el mecanismo LIFO (Last-In-First-Out) para almacenar los datos. El último elemento insertado en la pila es el primero que se elimina. Sus operaciones comunes son: crear la pila, hacer *push* de un elemento, hacer *pop* de un elemento, encontrar el elemento superior (top) de la pila, encontrar cuántos elementos hay en la pila, etc.

Mientras se define un ADT, **no** hay que preocuparse demasiado por los detalles de implementación; estos solo entran en juego cuando realmente se usan. Diferentes tipos de ADTs se adecuan a distintos tipos de aplicaciones, y algunos están altamente especializados para tareas específicas.

## 1.5 ¿Qué es un algoritmo?

Consideremos el problema de **preparar una tortilla (omelette)**. Para prepararla, seguimos los pasos que se detallan a continuación:

1. Conseguir la sartén.  
2. Conseguir el aceite.  
   1. ¿Tenemos aceite?
      - Si sí, ponerlo en la sartén.
      - Si no, ¿queremos comprar aceite?
         - Si sí, entonces sal y cómpralo.
         - Si no, podemos terminar (abortar).  
3. Encender la estufa, etc.

Lo que estamos haciendo es, para un problema dado (preparar una tortilla), dar un procedimiento **paso a paso** para resolverlo. La definición formal de un algoritmo se puede expresar así:

> **Un algoritmo es un conjunto de instrucciones paso a paso para resolver un problema dado.**

Nota: No es necesario probar cada paso del algoritmo.

---

## 1.6 ¿Por qué analizar algoritmos?

Para ir de la ciudad "A" a la ciudad "B", existen muchas formas de lograrlo: en avión, autobús, tren o bicicleta. Dependiendo de la disponibilidad y conveniencia, elegimos la que más nos conviene. Del mismo modo, en informática, **existen múltiples algoritmos** para resolver el mismo problema (por ejemplo, para el problema de ordenamiento hay muchos algoritmos, como Insertion Sort, Selection Sort, Quick Sort, etc.). 

El análisis de algoritmos nos ayuda a determinar **cuál es el más eficiente** en términos de tiempo y espacio consumidos.

---

## 1.7 Objetivo del análisis de algoritmos

El objetivo del análisis de algoritmos es **comparar** algoritmos (o soluciones) principalmente en términos del **tiempo de ejecución**, pero también en otros factores (por ejemplo, memoria, esfuerzo de desarrollo, etc.).

---

## 1.8 ¿Qué es el análisis del tiempo de ejecución?

Es el proceso de determinar **cómo** aumenta el tiempo de procesamiento a medida que **aumenta el tamaño del problema** (o tamaño de la entrada). El tamaño de la entrada es el número de elementos de la entrada y, dependiendo del tipo de problema, puede ser de distinta naturaleza. Algunos ejemplos comunes de tipos de entrada son:

- Tamaño de un arreglo.
- Grado de un polinomio.
- Número de elementos en una matriz.
- Número de bits en la representación binaria de la entrada.
- Vértices y aristas en un grafo.

---

## 1.9 ¿Cómo comparar algoritmos?

Para comparar algoritmos, definamos algunas **medidas objetivas**:

- **¿Tiempos de ejecución?** No es una buena medida, porque los tiempos de ejecución dependen de la computadora específica.  
- **¿Número de sentencias ejecutadas?** Tampoco es una buena medida, ya que varía según el lenguaje de programación y el estilo del programador.  
- **Solución ideal**: Expresar el tiempo de ejecución de un algoritmo como una función del tamaño de la entrada $n$, es decir, $f(n)$ y comparar dichas funciones. Este método es independiente del tiempo de máquina, del estilo de programación, etc.

---

## 1.10 ¿Qué es la tasa de crecimiento?

La tasa a la que aumenta el tiempo de ejecución como función de la entrada se llama **tasa de crecimiento** (Rate of Growth). Imagina que vas a una tienda a comprar un automóvil y una bicicleta. Si un amigo te ve y te pregunta qué estás comprando, normalmente contestarás "un auto". Esto se debe a que el costo del auto es mucho mayor que el de la bicicleta (aproximando el costo de la bicicleta al del auto).

$$
\text{Costo Total} = \text{costo\_auto} + \text{costo\_bicicleta} \\
\text{Costo Total} \approx \text{costo\_auto}
$$

Para grandes valores de $n$, podemos **ignorar los términos de menor orden**. Por ejemplo, si tenemos $n^4 + 2n^2 + 100n + 500$, nos aproximamos a $n^4$ para valores grandes de $n$, ya que $n^4$ domina los demás términos.

---

## 1.11 Tasas de crecimiento más comunes

Se muestra la relación entre diferentes tasas de crecimiento, desde las más lentas hasta las más rápidas. Algunas de las más utilizadas son:

- $O(1)$: Constante  
- $O(\log n)$: Logarítmica  
- $O(n)$: Lineal  
- $O(n \log n)$: Lineal-logarítmica  
- $(O(n^2)$: Cuadrática  
- $O(n^3)$: Cúbica  
- $O(2^n)$: Exponencial  

Como ejemplo, el tiempo para **agregar un elemento al frente de una lista enlazada** es $O(1)$. El tiempo para **encontrar un elemento en un array desordenado** es $O(n)$. El algoritmo de **Merge Sort** es $O(n \log n)$, etc.


## 1.12 Tipos de análisis

Para analizar un algoritmo, debemos saber con qué entradas el algoritmo tarda menos tiempo (buen rendimiento) y con qué entradas tarda más tiempo (peor rendimiento). Como ya hemos visto, un algoritmo se puede expresar en forma de una función. Eso significa que podemos representarlo con **múltiples expresiones**: una para el caso en que tarda menos y otra para el caso en que tarda más.

En general:

- **Peor caso (worst case)**: Define la entrada con la cual el algoritmo tarda más tiempo (más lento).  
- **Mejor caso (best case)**: Define la entrada con la cual el algoritmo tarda menos tiempo (más rápido).  
- **Caso promedio (average case)**: Da una **predicción** sobre el tiempo de ejecución asumiendo que la entrada es aleatoria.

$$\text{Límite inferior} <= \text{Tiempo medio} <= \text{Límite superior}$$

Así, existen **tres** tipos de análisis: **peor caso**, **mejor caso** y **caso promedio**. Cada uno puede expresarse como una función de $n$. Por ejemplo:

$$
f(n) = n^2 + 500 \quad (\text{peor caso})
$$
$$
f(n) = n + 100n + 500 \quad (\text{mejor caso})
$$

Para el caso promedio, se haría algo similar. 



###  Implementación y comparación de algoritmos en Python


````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
````

````python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr
````

Para comparar el rendimiento de estos algoritmos, se puede utilizar el módulo timeit para medir el tiempo de ejecución de cada uno.

````python
import timeit
import random

def generar_lista(n):
    return [random.randint(0, n) for _ in range(n)]

def medir_tiempo(algoritmo, arr):
    inicio = timeit.default_timer()
    algoritmo(arr.copy())
    fin = timeit.default_timer()
    return fin - inicio

n = 1000
lista = generar_lista(n)

tiempo_bubble = medir_tiempo(bubble_sort, lista)
tiempo_merge = medir_tiempo(merge_sort, lista)

print(f"Tiempo Bubble Sort: {tiempo_bubble:.6f} segundos")
print(f"Tiempo Merge Sort: {tiempo_merge:.6f} segundos")
````

Para visualizar la diferencia en el rendimiento, se pueden graficar los tiempos de ejecución para diferentes tamaños de entrada.

````python
import matplotlib.pyplot as plt

def comparar_algoritmos(tamanos):
    tiempos_bubble = []
    tiempos_merge = []
    for n in tamanos:
        lista = generar_lista(n)
        tiempo_bubble = medir_tiempo(bubble_sort, lista)
        tiempo_merge = medir_tiempo(merge_sort, lista)
        tiempos_bubble.append(tiempo_bubble)
        tiempos_merge.append(tiempo_merge)
    return tiempos_bubble, tiempos_merge

tamanos = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
tiempos_bubble, tiempos_merge = comparar_algoritmos(tamanos)

plt.plot(tamanos, tiempos_bubble, label='Bubble Sort')
plt.plot(tamanos, tiempos_merge, label='Merge Sort')
plt.xlabel('Tamaño de la lista')
plt.ylabel('Tiempo de ejecución (s)')
plt.title('Comparación de Algoritmos de Ordenamiento')
plt.legend()
plt.show()
````

## Aplicación ¿Qué es el ordenamiento (Sorting)?

El ordenamiento es un algoritmo que **organiza los elementos de una lista** en un cierto orden (ya sea ascendente o descendente). La salida es una **permutación** o **reordenamiento** de la entrada.


## ¿Por qué es necesario el ordenamiento?

El ordenamiento es una de las categorías más importantes de algoritmos y se ha investigado mucho al respecto. Ordenar puede **reducir significativamente la complejidad** de un problema y se emplea con frecuencia en algoritmos de bases de datos y búsquedas.

## Clasificación de los algoritmos de ordenamiento

Los algoritmos de ordenamiento suelen clasificarse según distintos criterios:

### Según el número de comparaciones

En este método, los algoritmos de ordenamiento se clasifican según la cantidad de **comparaciones**. Para los algoritmos de ordenamiento basados en comparaciones, el comportamiento en el **mejor caso** suele ser $O(n \log n)$ y en el **peor caso** $O(n^2)$. Estos algoritmos evalúan los elementos de la lista mediante operaciones de comparación de claves y, para la mayoría de entradas, necesitan al menos $O(n \log n)$ comparaciones.

Más adelante se discutirán algunos algoritmos de ordenamiento **no basado en comparaciones** (ordenamiento lineal) como **Counting sort**, **Bucket sort**, **Radix sort**, etc. Dichos algoritmos imponen ciertas restricciones sobre la entrada para mejorar la complejidad.

### Según el número de intercambios (swaps)

Aquí, los algoritmos de ordenamiento se categorizan según la **cantidad de intercambios** (también llamados **inversiones**).

### Según el uso de memoria

Algunos algoritmos de ordenamiento son “in-place” (en el lugar), y necesitan $O(1)$ o $O(\log n)$ memoria adicional para ubicar temporalmente los datos durante la ordenación.

### Según la recursividad

Los algoritmos de ordenamiento pueden ser **recursivos** (por ejemplo, Quick sort) o **no recursivos** (por ejemplo, Selection sort e Insertion sort). También hay algoritmos que usan ambos enfoques (Merge sort).

### Según la estabilidad

Un **algoritmo de ordenamiento es estable** si, para todos los índices $i$ y $j$ tales que la clave de $A[i]$ es igual a la de $A[j]$, si el registro $R[i]$ precede a $R[j]$ en el archivo original, entonces $R[i]$ precede a $R[j]$ en la lista ordenada. 

Algunos algoritmos mantienen el orden relativo de los elementos con claves iguales (los elementos equivalentes retienen su posición relativa incluso después de ordenar).

### Según la adaptabilidad

Con ciertos algoritmos, la complejidad **cambia** si la entrada está **casi ordenada** (por ejemplo, Quick sort): la “pre-ordenación” de la entrada afecta el tiempo de ejecución. Estos algoritmos que **se adaptan** se conocen como **adaptativos**.

---

## Otras clasificaciones

Existen más métodos de clasificación de los algoritmos de ordenamiento:

1. **Ordenamiento interno (Internal Sort)**  
   Son algoritmos que usan **exclusivamente la memoria principal** durante el proceso. Asumen acceso aleatorio de alta velocidad a toda la memoria.

2. **Ordenamiento externo (External Sort)**  
   Son algoritmos que usan **memoria externa** (cinta, disco, etc.) durante el ordenamiento.

---

## Bubble Sort

**Bubble sort** es el algoritmo de ordenación más simple. Funciona iterando sobre el arreglo de entrada desde el primer elemento hasta el último, **comparando cada par** de elementos e **intercambiándolos** si es necesario. Continúa iterando hasta que **no** se requieran más intercambios. Obtiene su nombre de la forma en que los elementos más pequeños “suben como burbujas” hacia el principio de la lista.  

En general, **Insertion sort** suele tener mejor rendimiento que **Bubble sort**. Algunos investigadores sugieren que ni siquiera se enseñe Bubble Sort debido a su sencillez y su elevada complejidad temporal.

La única ventaja significativa de Bubble Sort sobre otros algoritmos es que **puede detectar** si la lista ya está ordenada.

### Implementación

```python
def BubbleSort(A):
    for i in range(len(A)):
        for k in range(len(A) - 1, i, -1):
            if A[k] < A[k - 1]:
                swap(A, k, k - 1)

def swap(A, x, y):
    temp = A[x]
    A[x] = A[y]
    A[y] = temp

A = [534, 246, 933, 127, 277, 321, 454, 565, 220]
BubbleSort(A)
print(A)
```

Este algoritmo emplea $O(n^2)$ (incluso en el **mejor caso**). Podemos mejorarlo con una bandera (flag). Si no hay más intercambios, el algoritmo se detiene. Si la lista **ya está ordenada**, esta versión “mejorada” se detiene antes y reduce el mejor caso a $O(n)$.

```python
def BubbleSort(A):
    for i in range(len(A)):
        swapped = 1
        for i in range(len(A)):
            if(swapped ==0):
                return
            for k in range(len(A)-1,i, -1):
                 if A[k] < A[k - 1]: 
                    swap(A, k, k - 1)  
                    swapped = 1 

def swap(A, x, y):
    temp = A[x]
    A[x] = A[y]
    A[y] = temp

A = [127, 220, 246, 277, 321, 454, 534, 565, 933]
BubbleSort(A)
print(A)
```

**Rendimiento**:  
- Peor caso: $O(n^2)$  
- Mejor caso (versión mejorada): $O(n)$

## Selection Sort

Selection sort es un algoritmo de ordenación **in-place**. Funciona bien con arreglos (o ficheros) pequeños. Se usa cuando los valores son muy grandes y las claves pequeñas, pues la selección se hace con base en las claves y solo intercambia cuando es necesario.

### Ventajas

- Fácil de implementar.  
- Ordenamiento in-place (no requiere espacio extra significativo).

### Desventajas

- No escala bien: $O(n^2)$.

### Algoritmo

1. Encuentra el valor mínimo en la lista.  
2. Intercámbialo con el valor de la posición actual.  
3. Repite este proceso para todos los elementos hasta que el arreglo esté ordenado.

```python
def SelectionSort(A):
    for i in range(len(A)):
        least = i
        for k in range(i+1, len(A)):
            if A[k] < A[least]:
                least = k
        swap(A, least, i)

def swap(A, x, y):
    temp = A[x]
    A[x] = A[y]
    A[y] = temp

A = [54, 26, 93, 17, 77, 31, 44, 55, 20]
SelectionSort(A)
print(A)
```

**Rendimiento**:  
- Peor caso: $O(n^2)$  
- Mejor caso: $O(n^2)$ (no mejora incluso si ya está casi ordenado).

## Insertion Sort

Insertion sort es un método simple y eficiente basado en comparaciones. En cada iteración se **extrae** un elemento de los datos de entrada y **se inserta** en la posición correcta dentro de la lista que se está ordenando. Este proceso se repite hasta que **todos** los elementos hayan sido ubicados.

### Ventajas

- Implementación sencilla.  
- Eficiente para datos de tamaño pequeño.  
- Adaptativo: si la lista está *casi ordenada*, puede rendir en $O(n + d)$, donde $d$ es el número de “inversiones”.  
- En la práctica, más eficiente que Selection y Bubble, pese a que su peor caso también sea $O(n^2)$.  
- **Estable**: mantiene el orden relativo de los datos de la lista si hay claves iguales.  
- **In-place**: solo necesita $O(1)$ espacio extra.  
- **Online**: puede ordenar la lista a medida que la recibe.

### Algoritmo

Cada repetición extrae un elemento del arreglo de entrada y lo coloca en la posición correcta de la sublista ya ordenada, hasta que no queden elementos de entrada.

```python
def InsertionSort(A):
    for i in range(1, len(A)):
        temp = A[i]
        k = i
        while k > 0 and temp < A[k-1]:
            A[k] = A[k-1]
            k -= 1
        A[k] = temp

A = [54, 26, 93, 17, 77, 31, 44, 55, 20]
InsertionSort(A)
print(A)
```

**Ejemplo**  
Dado un arreglo $[6, 8, 1, 4, 5, 3, 7, 2]$:

1. (Índice 0): 6 8 1 4 5 3 7 2  
2. (Índices 0-1): Se comprueba 6 y 8; nada que cambiar.  
3. (Índices 0-2): Insertar `1` en la posición correcta => 1 6 8 4 5 3 7 2  
4. (Índices 0-3): Insertar `4` en la posición correcta => 1 4 6 8 5 3 7 2  
5. Y así sucesivamente…

Al final queda ordenado: 1 2 3 4 5 6 7 8.

**Análisis**  

### **Análisis del peor caso:**

El peor caso ocurre cuando, para cada $i$, el bucle interno tiene que mover todos los elementos $A[1], A[2], \ldots, A[i - 1]$ (lo que sucede cuando $A[i]$ = clave es menor que todos ellos), lo que toma un tiempo de $\Theta(i - 1)$.

$$
T(n) = \Theta(1) + \Theta(2) + \Theta(2) + \ldots + \Theta(n - 1)
$$
$$
= \Theta(1 + 2 + 3 + \ldots + (n - 1)) = \Theta\left(\frac{n(n - 1)}{2}\right) \approx \Theta(n^2)
$$


### **Análisis del caso promedio:**

Para el caso promedio, el bucle interno insertará $A[i]$ en el medio de $A[1], A[2], \ldots, A[i - 1]$. Esto toma un tiempo de $\Theta(i/2)$.

$$
T(n) = \sum_{i=1}^{n} \Theta(i/2) \approx \Theta(n^2)
$$


### **Rendimiento:**

- **Complejidad del peor caso:** $O(n^2)$
- **Complejidad del mejor caso:** $O(n)$
- **Complejidad del caso promedio:** $O(n^2)$
- **Complejidad de espacio en el peor caso:** $O(n^2)$ total, $O(1)$ auxiliar. 
    



## **Comparación con otros algoritmos de ordenamiento**

El **ordenamiento por inserción (Insertion Sort)** es uno de los algoritmos de ordenamiento elementales con una complejidad de tiempo de $O(n^2)$ en el peor caso. El ordenamiento por inserción se utiliza cuando los datos están casi ordenados (debido a su capacidad adaptativa) o cuando el tamaño de entrada es pequeño (debido a su bajo costo de sobrecarga). Por estas razones, y debido a su estabilidad, el ordenamiento por inserción se usa como caso base recursivo (cuando el tamaño del problema es pequeño) en algoritmos de ordenamiento de "divide y vencerás" con alta sobrecarga, como el **merge sort** o el **quick sort**.


### **Notas:**

1. **Bubble Sort** realiza $\frac{n^2}{2}$ comparaciones y $\frac{n^2}{2}$ intercambios (inversiones) tanto en el caso promedio como en el peor caso.
2. **Selection Sort** realiza $\frac{n^2}{2}$ comparaciones y $n$ intercambios.
3. **Insertion Sort** realiza $\frac{n^2}{4}$ comparaciones y $\frac{n^2}{8}$ intercambios en el caso promedio, y en el peor caso estas cifras se duplican.
4. **Insertion Sort** es casi lineal para entradas parcialmente ordenadas.
5. **Selection Sort** es más adecuado para elementos con valores grandes y claves pequeñas.

#### **Pregunta 1**

**Análisis del costo temporal de la función `func(n)`**

La función `func(n)` realiza una suma de los cuadrados de los números del 0 al `n-1`. Analicemos cada operación en términos de costos temporales dados:

```python
def func(n):
    sum = 0  # costo temporal de la asignación = c1
    # se asume que la inicialización de i a 0 al inicio del bucle tiene costo 0
    for i in range(n):  # costo por cada incremento de i y verificación del rango = c2

        # costo de la asignación = c1

```

Desglosemos el costo total:

1. **Inicialización:**
   - `sum = 0` → costo = `c1`

2. **Bucle `for`:** Se ejecuta `n` veces.
   - **Por cada iteración:**
     - Verificación del rango e incremento de `i` → costo = `c2`
     - Cálculo de `i * i` → costo = `c4`
     - Suma `sum + (i * i)` → costo = `c3`

- Inicialización: `c1`
- Bucle: `n * (c2 + c4 + c3 + c1)`
- Retorno: `c5`

Por lo tanto, el costo total es:
$$
c1 + n \times (c2 + c3 + c4 + c1) + c5
$$

$$

$$

**Costo total en el peor caso de ejecutar *Insertion Sort* en un arreglo ordenado en orden descendente**

El pseudocódigo proporcionado es:

```plaintext

    key = A[i]  # costo de asignación = c1
    j = i - 1   # costo de asignación = c1
    while j >= 0 and A[j] > key:  # costo de comparación = c2

        j = j - 1  # costo de decremento y asignación = c1
    A[j + 1] = key  # costo de asignación = c1
```

**Análisis del peor caso:**

En el peor caso, el arreglo está ordenado en orden descendente. Para cada elemento `A[i]`, se deben comparar y desplazar todos los elementos de la porción ordenada (de `A[0]` a `A[i-1]`).

1. **Por cada iteración `i`:**
   - Asignaciones iniciales: `key = A[i]` y `j = i - 1` → costo = `2 * c1`
   - **Bucle `while`:**
     - Se realizan `i` comparaciones (`A[j] > key`).
     - Se realizan `i` asignaciones para desplazar los elementos (`A[j + 1] = A[j]`).
     - Se realizan `i` decrementos y asignaciones para `j = j - 1` → costo = `i * (c1 + c2)`
   - Asignación final: `A[j + 1] = key` → costo = `c1`

2. **Costo total por iteración `i`:** `2 * c1 + i * (c1 + c2) + c1 = 3 * c1 + i * (c1 + c2)`

3. **Costo total para todas las iteraciones:**
   $$
   \sum_{i=1}^{n-1} (3c1 + i(c1 + c2)) = 3c1(n - 1) + (c1 + c2) \sum_{i=1}^{n-1} i
   $$
   $$
   = 3c1(n - 1) + (c1 + c2) \cdot \frac{n(n - 1)}{2}
   $$

**Comparación con las opciones:**

- **Opción 2:**
  $$
  c1 \times n + \frac{n(n - 1)}{2} \times (c1 + c2)
  $$
  Coincide con nuestro análisis, ya que los términos adicionales son constantes y no afectan la expresión principal.

- **Opción 3:** $\Theta(n^2)$ es la notación asintótica que representa el comportamiento cuadrático.

Las opciones correctas son la **Opción 2** y la **Opción 3**.

#### **Pregunta 6**

**Análisis de *Insertion Sort* en un arreglo casi ordenado con $k$ elementos desordenados**

Dado:
- Tamaño del arreglo: $n = 10,000$
- Número de elementos desordenados: $k = 10$

**Parte 1: Complejidad temporal en función de $n$ y $k$**

En un arreglo casi ordenado donde solo $k$ elementos están desordenados, el algoritmo de *Insertion Sort* realizará:

- **Para los elementos ya ordenados:** Comparaciones mínimas, principalmente constantes.
- **Para los $k$ elementos desordenados:** Cada uno requerirá, en el peor caso, desplazar hasta $k$ elementos hacia la posición correcta.

Por lo tanto, la complejidad temporal es:

$$
O(n + k \cdot n)
$$

Sin embargo, dado que $k$ es significativamente menor que $n$, y generalmente considerado como constante, la complejidad se aproxima a:

$$
O(n + k^2) \approx O(n)
$$

**Respuesta parte 1:** **La complejidad temporal es $O(n + k^2)$, que para $k$ pequeño se aproxima a $O(n)$.**

**Parte 2: Comportamiento cuando $k$ permanece constante y $n$ crece**

Si $k$ permanece constante mientras $n$ crece:

- La parte dominante de la complejidad es $O(n)$.
- Los términos relacionados con $k$ no afectan significativamente el crecimiento asintótico.

**Respuesta parte 2:** **El tiempo de ejecución crece linealmente con $n$, es decir, permanece en $O(n)$.**

**Parte 3: Explicación de por qué la complejidad se aproxima a $O(n)$ en este caso**

Cuando solo $k$ elementos están desordenados en un arreglo de tamaño $n$, la mayoría de las inserciones en *Insertion Sort* requieren tiempo constante (ya que los elementos están ya en orden). Solo los $k$ elementos desordenados necesitan desplazamientos adicionales, pero si $k$ es constante, estos desplazamientos no afectan el crecimiento asintótico con respecto a $n$.

Por lo tanto, la mayor parte del trabajo realizado por el algoritmo es lineal en relación con $n$, y las operaciones adicionales para los $k$ elementos desordenados no incrementan la complejidad más allá de $O(n)$.

**Respuesta parte 3:** **Porque solo una cantidad constante de elementos requiere operaciones adicionales, el algoritmo realiza principalmente operaciones lineales respecto a $n$, lo que hace que la complejidad se aproxime a $O(n)$ en lugar de $O(n^2)$.**
